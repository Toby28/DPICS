```python
from moviepy.editor import VideoFileClip

def cut_video(source_path, start_time, end_time, output_path_1, output_path_2):
    """
    Cuts a clip from a video between start_time and end_time.
    
    Args:
    source_path (str): Path to the source video file.
    start_time (int or float): Start time in seconds.
    end_time (int or float): End time in seconds.
    output_path (str): Path where the cut video will be saved.
    """
    # Load the video file
    video = VideoFileClip(source_path)
    print(video)
    
    # Cut the video
    cut_clip = video.subclip(start_time, end_time)
    
    # Write the result to the output file
    #cut_clip.write_videofile(output_path, codec='libx264')
    cut_clip.write_gif(output_path_1)
    cut_clip.write_videofile(output_path_2)

# Example usage

#sourcepath="C:/Users/chlin/Downloads/2022003P0120220331mp4_20220520T163101Z.mp4"
sourcepath="C:/Users/chlin/Downloads/2022003C0320220405mp4_20220520T163024Z.mp4"

start_time=7*60 + 8.0
end_time= 7*60+ 13.0

outputpath_1="test9.gif"
outputpath_2="movie9.mp4"

cut_video(sourcepath, start_time, end_time, outputpath_1, outputpath_2)
```
